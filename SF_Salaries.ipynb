```python
#Replace 'Not Provided' in EmployeeName' Column to NaN
df['BasePay']=df['BasePay'].replace('Not Provided',np.nan)
```

```python
df['BasePay']
```

```python
df['BasePay']=pd.to_numeric(df['BasePay'],errors='coerce')
```

```python
df['BasePay'].describe()
```

```python
# Replace 'Not Provided' in EmployeeName' Column to NaN
df['EmployeeName']=df['EmployeeName'].replace('Not provided',np.nan)
```

```python
df['EmployeeName']
```

```python
df.columns
```

```python
#Drop The Rows Having 5 Missing Values
df.drop(df[df.isnull().sum(axis=1)==5].index,axis=0,inplace=True)#inplace true makes changes in original dataframes
```

```python
df.columns
```

```python
#Find Job Title of ALBERT PARDINI
df[df['EmployeeName']=='ALBERT PARDINI']['JobTitle']
```

```python
#How Much ALBERT PARDINI Make (Include Benefits)?
df[df['EmployeeName']=='ALBERT PARDINI']['TotalPayBenefits']
```

```python
#Display Name of The Person Having The Highest BasePay
df[df['BasePay'].max()==df['BasePay']]['EmployeeName']
```
